```python
# Get LIST of all files and directories in a specified directory
# or current working directory
for f in sorted(os.listdir()):
  # First string is root of path (everything before last period)
  # Second string is extension (everything after the last period, INCLUDING period itself)
  filename, file_ext = os.path.splitext(f)
  print(f'{filename}/{file_ext}')

  # Default separator is any whitespace
  planet, course, num = filename.split('-')
  print(f'{planet}/{course}/{num}')

  # Remove leading and trailing characters, default: whitespace (spaces, tabs, newlines)
  planet = planet.strip()
  course = course.strip()
  num = num.strip()

  course = course.replace(' ', '_')
  print(f'{planet}/{course}/{num}')

  print('{:02}_{}_{}{}'.format(int(num[1:]), course, planet, file_ext))
  print()
  new_name = '{:02}_{}_{}{}'.format(int(num[1:]), course, planet, file_ext)
  os.rename(f, new_name)
```
