Assume as starting point, we are in an empty `basics` folder

We can check where we are

    pwd

and we have

    /Users/jianjing/Desktop/basics

#### Create a directory

    mkdir TestDir

we can see via

    ls -l

that in the `basics`, we now have

    total 0
    drwxr-xr-x  2 jianjing  staff  64 Jun 19 16:09 TestDir

#### Create a file

We first navigate into it

    cd TestDir/

and create a file

    touch test_file.txt

with

    ls -l

we have

    total 0
    -rw-r--r--  1 jianjing  staff  0 Jun 19 16:11 test_file.txt

#### Open file and edit

The following will open an text editor

    open test_file.txt

#### Copy file

We do

    cp test_file.txt copy_file.txt

via

    ls -l

we can see the copied file|

    total 16
    -rw-r--r--@ 1 jianjing  staff  13 Jun 19 16:13 copy_file.txt
    -rw-r--r--@ 1 jianjing  staff  13 Jun 19 16:12 test_file.txt

#### Rename/move a file

These two function uses same `mv` command

    mv test_file.txt orig_file.txt

with

    ls -l

we can see the renamed file

    -rw-r--r--@ 1 jianjing  staff  13 Jun 19 16:13 copy_file.txt
    -rw-r--r--@ 1 jianjing  staff  13 Jun 19 16:12 orig_file.txt

To `move` a file, we leave out file name and replace with a location

We first create a new folder `SubDir1`

    mkdir SubDir1

in the `TestDir`, we now have

    copy_file.txt	orig_file.txt	SubDir1

To move

    mv orig_file.txt SubDir1/

and in `TestDir` we now have

    copy_file.txt	SubDir1

we move to `SubDir1`

    cd SubDir1

and we have

    orig_file.txt

We can do move and rename at same time

    mv orig_file.txt ../test_file.txt

we go back to `TestDir`

    cd ..

and we have

    copy_file.txt	SubDir1		test_file.txt

#### Delete a file

    rm copy_file.txt

and in `TestDir`, this copy_file is gone

    SubDir1		test_file.txt

#### Modify directory

First, let's make a copy of TestDir to CopyDir

We go back to `basics`

    cd ..

we can copy using (with -R flag), either way seems to work

    cp -R TestDir/ CopyDir/
    cp -R TestDir/ CopyDir

#### Rename/move directory

For the purpose of deciding “is this a rename or a move-into-directory?”

`mv` first looks at the destination, not at whether the source is a file or directory
  - if target path exists and is a directory, then move
  - if not, then rename

It will, of course, verify that the source exists, but its type isn't what drives the algorithm at that stage.

    mv TestDir/ OrigDir/  # OrigDir/ does not exist, so rename
    mv OrigDir/ CopyDir/  # CopyDir/ exists, so move
    cd CopyDir/
    mv OrigDir/ ../TestDir/  # ../TestDir/ move and rename

#### Delete directory

    rm -R CopyDir/
