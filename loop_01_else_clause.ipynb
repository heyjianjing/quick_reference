#### For/else loop

In a `for/else` loop, both can be executed, unlike `if/else` condition

```python
my_list = [1, 2, 3, 4, 5]

for i in my_list:
  print(i)
else:
  print("Hit the for/else statement")
```

Semantically, it should be called `no break`

First, we know how `break` works

```python
for i in my_list:
  print(i)
  if i == 3:
    break
else:
  print("Hit the for/else statement")
```

Now, we can think `else` as saying

If no break is hit, then print what is under `else`

```python
for i in my_list:
  print(i)
  if i == 9:
    break
else:
  print("Hit the for/else statement")
```

#### While/else loop

Same for while/else loop

```python
i = 1
while i < 5:
  print(i)
  i += 1
else:
  print("Hit the while/else statement")
```

```python
i = 1
while i < 5:
  print(i)
  i += 1
  if i == 3:
    break
else:
  print("Hit the while/else statement")
```

#### Practice use

```python
def find_index(to_search, target):
  for i, value in enumerate(to_search):
    if value == target:
      break
  else:
    return -1
  return i

my_list = ['a', 'b', 'c']
print(find_index(my_list, 'a'))
print(find_index(my_list, 'd'))
```
