#### Create and save a python script

```python
%%writefile first_module.py
print(f"First module's name: {__name__}")
```

#### Run python script directly

```python
!python first_module.py
```

What is going on here?

Before python runs any of the actual code, it sets a few special variables `corresponding to the script`, `__name__` is one of them

When Python run a script `directly`, it sets variable `__name__` to value `'__main__'`

#### Import python script

```python
%%writefile second_module.py
import first_module
print(f"Second module's name: {__name__}")
```

```python
!python second_module.py
```

When a module is imported into a second script, the code of the module is executed

In this case, however, the `__name__` variable of the module being imported is set to the `file name` of the module

And the `__name__` variable of second script, which is now being ran directly, is set to `'__main__'`

#### if `__name__` == `'__main__'`

```python
%%writefile first_module.py

def main():
  print(f"First module's name: {__name__}")

# This basically check whether this py file is raw directly (condition is True)
# or is it being imported (condition is False)
if __name__ == '__main__':
  main()
```

If we run it `directly`, then `main()` will be executed since the condition is True

```python
!python first_module.py
```

If we run it by `importing` into second_module.py, then `main()` would `not` be executed, since the condition is False

```python
!python second_module.py
```

#### To see this more explicitly

```python
%%writefile first_module.py

if __name__ == '__main__':
  print('First module is ran directly')
else:
  print('First module is ran through import')
```
