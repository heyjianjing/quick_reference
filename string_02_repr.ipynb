#### Unambiguous vs readable

General idea

for `__repr__()` is to be `unambiguous`

for `__str__()` is to be `readable`

```python
import datetime
import pytz

a = datetime.datetime.utcnow().replace(tzinfo=pytz.UTC)

b = str(a)

print(f'str(a): {str(a)}')
print(f'str(b): {str(b)}')

print()

print(f'repr(a): {repr(a)}')
print(f'repr(b): {repr(b)}')

print()
```

Conceptually, what unambiguous means is that we can evaluate the output of `__repr__()` directly using python `eval()`

This can be useful for debugging, since when you just print out string `str()`, you are not going to tell whether it is `datetime.datetime`, or just a `string`

We you do `repr()`, it clearly tells whether it is `datetime.datetime`, or just a `string`

```python
a_1 = datetime.datetime(2025, 6, 17, 15, 50, 30, 223688, tzinfo=pytz.UTC)
b_1 = '2025-06-17 15:52:04.974673+00:00'
```

```python
type(a_1), type(b_1)
```

However, you cannot do things like `a_1 = 2025-06-17 15:52:04.974673+00:00`
