```python
from collections import namedtuple

# Regular tuple, less readable
color_tuple = (55, 155, 255)

# Dictionary, mutable
color_dict = {'red': 55, 'green': 155, 'blue': 255}

# Named tuple, immutable, combine best of reg tuple and dictionary
# First arg is class name
Color = namedtuple('Color', ['red', 'green', 'blue'])
white = Color(255, 255, 255)
```

```python
# Namedtuple can still be used like regular tuple
print(white[0])
```

```python
# Or we can use dictionary key style
print(white.blue)
```

```python
print(white)
```

```python
# and it is immutable
white.red = 100
```
