#### List

List hold entire result in memory

```python
def square_numbers(nums):
  result = []
  for i in nums:
    result.append(i**2)
  return result
```

```python
list_nums = [1,2,3,4,5]
```

```python
my_nums = square_numbers(list_nums)
```

```python
print(my_nums)
```

#### Generator

Generators don't hold entire result in memory

```python
def gen_square_numbers(nums):
  for i in nums:
    yield (i ** 2)
```

```python
my_gen = gen_square_numbers(list_nums)
```

```python
# Here generator is a placeholder
# No result has been computed
print(my_gen)
```

It yields one result at a time when we ask it for result

We can still use for loop to ask for result

```python
for num in my_gen:
  print(num)
```

#### Use list comprehension with `( )`

```python
my_gen = (x**2 for x in list_nums)
```

```python
print(my_gen)
```

```python
for num in my_gen:
  print(num)
```

#### Convert to list

```python
# you can convert generator to list
my_gen = (x**2 for x in list_nums)
```

```python
print(list(my_gen))
```
