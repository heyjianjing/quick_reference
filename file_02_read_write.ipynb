#### Read

Print out everything

```python
with open('test.txt', 'r') as f:
  print(f.read())
```

```python
with open('test.txt', 'r') as f:
  print(f.readlines())
```

Line by line

```python
with open('test.txt', 'r') as f:
  print(f.name)
  # read line by line
  f_content = f.readline()
  print(f_content, end='')

  f_content = f.readline()
  print(f_content, end='')
```

```python
with open('test.txt', 'r') as f:
  # iterate each line in file, one by one
  for line in f:
    print(line, end='')
```

```python
# read by size

with open('test.txt', 'r') as f:
  size_to_read = 20

  f_content = f.read(size_to_read)

  while len(f_content) > 0:
    print(f_content, end='*')
    f_content = f.read(size_to_read) # don't forget this line to avoid infinite loop
```

#### Write

```python
with open('test2.txt', 'w') as f:
  f.write('Test')
  f.seek(0)
  f.write('R')
```

#### Read and write to make a copy

```python
with open('test.txt', 'r') as rf:
  with open('test_copy.txt', 'w') as wf:
    for line in rf:
      wf.write(line)
```

#### Image data

```python
# !wget -O corgi.jpg https://blog.healthypawspetinsurance.com/wp-content/uploads/2021/09/brown-and-white-corgi-dog-outside.jpg
```

```python
# as "binary" mode

with open('corgi.jpg', 'rb') as rf:
  with open('corgi_copy.jpg', 'wb') as wf:
    for line in rf:
      wf.write(line)
```

```python
# we can also do chunk

with open('corgi.jpg', 'rb') as rf:
  with open('corgi_copy_chunk.jpg', 'wb') as wf:
    chunk_size = 4096
    rf_chunk = rf.read(chunk_size)
    while len(rf_chunk) > 0:
      wf.write(rf_chunk)
      rf_chunk = rf.read(chunk_size) # don't forget this line
```
