#### Navigation

    ctrl + a  # move to `beginning` of line
    ctrl + e  # move to `end` of line
    option + left/right  # move `one sequence of string` at a time (separated by whitespace)
    option + click  # move to `cursor location`

#### Edit

    ctrl + u  # remove all
    ctrl + w  # remove `before cursor`, one sequence of string at a time
    ctrl + k  # remove all `after cursor`

#### Track

    history  # show history of all commands
    ! + number  # run command corresponding to that number from history
    ctrl + r  # search keywords in command

#### Clear screen

    ctrl + l  # clear screen (keep scroll-back)
    cmd + k  # clear all
