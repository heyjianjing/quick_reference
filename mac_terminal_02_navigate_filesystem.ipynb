#### Print working directory

    pwd

we see, for example

    /Users/jianjingzhang

#### Navitage to folder

    cd Music

we get prompt

    jianjingzhang@QXO-N4960057L4 Music %

We can verify using `pwd`

    /Users/jianjingzhang/Music

#### Navitage to home directory

We can simply do

    cd

or

    cd ~

we verify using `pwd`

    /Users/jianjingzhang

#### List names of files under directory

    ls

and we get (we see a Music folder)

    Music

We can also list `hidden` files

    ls -a

and we see there are some

    .		..		.localized	Music

We can also display information in `long form`

    ls -l

we have

(`total` tells the number of 1K blocks used by the files in directory)

    total 0
    drwxr-xr-x  4 jianjingzhang  staff  128 Dec 16  2024 Music

and to see `hidden` files

    ls -la

we have

    total 0
    drwx------+  4 jianjingzhang  staff   128 Dec 16  2024 .
    drwxr-x---+ 36 jianjingzhang  staff  1152 Jun 17 13:48 ..
    -rw-r--r--   1 jianjingzhang  staff     0 Dec 16  2024 .localized
    drwxr-xr-x   4 jianjingzhang  staff   128 Dec 16  2024 Music

We can also specify directory to list names of files, for example, from `/Users/jianjingzhang`

    ls -la Music\

and we still can get what is inside Music, even though we are currently not in it

    total 0
    drwx------+  4 jianjingzhang  staff   128 Dec 16  2024 .
    drwxr-x---+ 36 jianjingzhang  staff  1152 Jun 17 13:48 ..
    -rw-r--r--   1 jianjingzhang  staff     0 Dec 16  2024 .localized
    drwxr-xr-x   4 jianjingzhang  staff   128 Dec 16  2024 Music

#### Relative path

##### Dot

`.` is current directory

`..` is parent directory of current directory

For example

    pwd

we have

    /Users/jianjingzhang/Music

and if we want to go to parent directory, we do

    cd ..

and we are at (with `pwd`)

    /Users/jianjingzhang

We can also `chain` multiple dot

From /Users/jianjingzhang/Music, we can get to Users/ via

    cd ../..

using `pwd`, we can verify

    /Users

##### Tilde

Same can be done with home directory `~`, for example, if we want to list filenames in home directory, we do

    ls ~

and we get

    Applications	Downloads	Music		Public
    Desktop		Library		Pictures	src
    Documents	Movies		projects

We can also get into folder in home directory

    cd ~/Pictures

with `pwd`, we have

    /Users/jianjingzhang/Pictures

Or get list of file names in that folder

    ls -la ~/Pictures

we have
